### 1. define Joint Probability 
Joint probability is the probability that **tow events occur together**.

Written as:

### 2. What is Marginal Probanility?
Marginal probaility is tha probabilty of **Only one event**, ignoring others.

It is calculated using **row or colum totals** of a table.

---

### 3. Why Contingency Tables are Useful?

Contingency tables help to:

- Show relationship between tow variables
- Calculate joint probility
- calculate marginal probability
- compare groups
- Find patterns

They are mainly used for **categorical data**.



```python
import pandas as pd
import matplotlib.pyplot as plt

#Creating contingency table data
data = {
    "Gender": ["Male", "Male", "Female", "Female"],
    "Result": ["Pass", "Fail", "Pass", "Fail"],
    "Count": [30, 10, 40, 20]
}

df = pd.DataFrame(data)
df
```

## ðŸ“Œ Contingency Table

```python
table = pd.pivot_table(df, values="Count", index="Gender", columns="Result", aggfunc="sum", margins=True)
table
```

## ðŸ“Œ Joint probability Example

Probability of Female AND Pass


```python
total = table.loc["All", "All"]
print("Total:", total)

joint_prob = table.loc["Female", "Pass"] / total
print("Joint Probability (Female and Pass):", joint_prob)   
```

## ðŸ“Œ Marginal Probability Examples


```python
p_female = table.loc["Female", "All"] / total
p_pass = table.loc["All", "Pass"] / total
p_female, p_pass
```

## âœ… Concusion

- Joint Probability -> Tow events together
- Marginal Probaility -> Single event
- Contingency Table -> Show relationship between veriables

These concepts are widely used in statistics and data analysis.

# ðŸ“Œ Mini Assignment
## Student Pass/Fail vs Gender - Probability analysis

Taske:
- Create dataset
- Contingency table
- Joint & Marginal probability
- venn diagram





```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib_venn import venn2

data = {
    "Gender": ["Male","Male","Male","Male","Female","Female","Female","Female","Female","Male"],
    "Result": ["Pass","Fail","Pass","Fail","Pass","Fail","Pass","Fail","Pass","Pass"],
}

df = pd.DataFrame(data)
df
```

## ðŸ“Œ Contingency Table (Gender VS Result)

```python
table = pd.crosstab(df["Gender"], df["Result"], margins=True)
table
```

## ðŸ“Œ Joint Probability (Female AND Pass)

```python
total = table.loc["All", "All"]

joint_prob = table.loc["Female", "Pass"] / total
joint_prob
```

## ðŸ“Œ Marginal Probability

```python
p_female = table.loc["Female", "All"] / total
p_pass = table.loc["All", "Pass"] / total
p_female, p_pass
```

## ðŸ“Œ Venn Diagram (Pass VS Female)

Circle A -> Female
Circle B -> Pass

```python
female = set(df[df["Gender"] == "Female"].index)
pass_ = set(df[df["Result"] == "Pass"].index)

venn2([female, pass_], set_labels = ("Female", "Pass"))
plt.title("Venn Diagram of Gender and Result")
plt.show()
```

## âœ… Conclusion

- joint probability shows two events together (Female AND Pass)
- Marginal probability showssingle event (only Female of only pass)
- Contingency table explains relationship
- venn diagram gives visual understanding
