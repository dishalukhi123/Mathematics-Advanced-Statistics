### 1. Define Descriptive Statistics
**Descriptive statistics** is the branch of statistics that is used to **summarize, organize, and present data** in a meaningful way using numbers, tables, and graphs.

**Example:** Mean, Median, Mode, Charts, Tables

---

### 2. When Whould Median Be Preferred Over Mean?
The **median** should ne preferred over the **mean** when:
- The data contains **outliers (extreme values)**
- The data is **skewed**
- The values are not evenly distributed

**Example:**
Income data, whare a few vary high values affect the mean.

### 3. Identify Measure of Dispersion
The following are **measures of dispersion**.

- **Range**
- **Variance**
- **Standard Deviation**

These measures describe **how spread out the data values are**.

---

### 4. Can Tow Datasets Have the Same Mean but Different Standard Deviation?
**Yes**, two datasets can have the **same mean** but **different standard deviation**

- Mean shows the **central value**
- Standerd deviation shows the **spread of data**

**Example:**
- Dataset A: 48, 50, 52
- Dataset B: 30, 50, 70
