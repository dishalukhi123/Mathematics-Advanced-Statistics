## 1. Define Statistics

Statistics is the branch of mathematics that deals with collecting, organizing, analyzing, and interpreting data to make decisions or conclusions.


## 2. Difference between Descriptive and Inferential Statistics

| Descriptive Statistics | Inferential Statistics |
|-----------------------|------------------------|
| It summarizes data | It predicts or concludes about a population |
| Uses tables, charts, mean, median, etc. | Uses probability, hypothesis testing |
| Deals with given data only | Uses sample data to make decisions |
| Example: Average marks of a class | Example: Predicting results of all students |


## 3. Identify Data Types (Numerical / Categorical)

- **Phone number** ‚Üí üìÇ *Categorical*  
  (Numbers are used only as labels, not for calculation)

- **Salary** ‚Üí üî¢ *Numerical*  
  (Used in calculations like average, total, etc.)

- **Department name** ‚Üí üìÇ *Categorical*  
  (Examples: HR, IT, Sales)


## 4. Why Visualization is Important?

Visualization is important because:

- It makes data easy to understand  
- Helps to identify patterns and trends  
- Saves time compared to reading raw data  
- Makes comparison clear and simple  

üìä **Examples:** Bar chart, Pie chart, Line graph


## 5. Difference between Population and Sample

| Population | Sample |
|-----------|--------|
| Entire group of data | Small part of population |
| More accurate | Less costly and faster |
| Hard to study fully | Easy to analyze |
| Example: All students in a college | Example: 50 students from the college |

‚ú® **Tip for exams:**  
Population = Whole  
Sample = Part

### üìä Student Dataset (Table Format)


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Roll No": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Marks": [85, 92, 78, 90, 88, 76, 95, 89, 84, 91],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
    "city": [
        "New York",
        "Los Angeles",
        "Chicago",
        "Houston",
        "Phoenix",
        "Philadelphia",
        "San Antonio",
        "San Diego",
        "Dallas",
        "San Jose",
    ]}
students_df = pd.DataFrame(data)
students_df
```

## 1. Identify Data Types

- Roll No ‚Üí Numerical  
- Marks ‚Üí Numerical  
- Gender ‚Üí Categorical  
- City ‚Üí Categorical  


## 2. Calculate average marks.


```python
avg_marks = students_df["Marks"].mean()
avg_marks
```

## 3. Create:
   - Bar chart (Gender)
   - Histogram (Marks)

```python
gender_counts =students_df["Gender"].value_counts()

plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Number of Students')            
plt.title('Number of Students by Gender')
plt.show()
```

```python
plt.hist(students_df["Marks"], bins=5, color='green', edgecolor='black')
plt.xlabel('Marks') 
plt.ylabel('Number of Students')
plt.title('Distribution of Marks')
plt.show()
```

## 4. Take a random sample of 5 students and compare average marks with population.

```python
sample_students = students_df.sample(n=5, random_state=1)
sample_students
```

```python
population_avg = students_df["Marks"].mean()
sample_avg = sample_students["Marks"].mean()

population_avg, sample_avg
```
