## Mini Assignment

**Claim**

Average service time = 5 minutes

**Hypotheses**

Hâ‚€: Î¼ = 5

Hâ‚: Î¼ â‰  5

**Sample (Assumed)**

Sample mean = 5.6

Sample size = 30

**Result**

p-value = 0.03

95% CI = (5.4 , 5.8)

**Final Conclusion (Write in exam):**

Since p-value < 0.05 and 5 is not inside confidence interval, we reject the null hypothesis.
Average service time is not 5 minutes.

```python
import numpy as np
from scipy.stats import norm

np.random.seed(42)
population_mean = 5
population_std = 1
sample_size = 30

sample_data = np.random.normal(5.6, population_std, sample_size)
sample_mean = np.mean(sample_data)

z_value = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_value)))

print('Sample Mean:', round(sample_mean,2))
print('Z:', round(z_value,3))
print('P-value:', round(p_value,4))
```

```python
z_critical = 1.96
margin = z_critical * (population_std / np.sqrt(sample_size))
lower = sample_mean - margin
upper = sample_mean + margin

print('95% CI:', round(lower,2), 'to', round(upper,2))
```

## Interpretation

If p < 0.05 reject H0.
If 5 not in CI reject H0.

Type I and Type II are possible errors.
