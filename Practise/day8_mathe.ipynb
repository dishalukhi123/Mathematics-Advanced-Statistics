## Difference between Descriptive and Inferential Statistics

**Descriptive Statistics:** Summarizes data using mean, median, mode, tables, and graphs.

**Inferential Statistics:** Uses sample data to make conclusions about a population using probability.

## Hypothesis Testing

Hypothesis testing is a statistical method used to decide whether sample data provides enough evidence to support or reject a claim about a population.

## Mini Assignment

**Claim:** Average delivery time is 30 minutes

Let μ be the average delivery time.

**Null Hypothesis (H₀):** μ = 30 minutes

**Alternative Hypothesis (H₁):** μ ≠ 30 minutes

Assume Sample:
**Sample size** = 25
**Sample mean** = 34 minutes

Interpretation:

If p-value < 0.05 → Reject H₀ → Delivery time ≠ 30

If p-value > 0.05 → Accept H₀ → Delivery time ≈ 30

```python
import numpy as np
from scipy.stats import norm

# Assume sample data
np.random.seed(42)
sample_size = 25
population_mean = 30
population_std = 6

sample_data = np.random.normal(34, population_std, sample_size)
sample_mean = np.mean(sample_data)

# One-sample Z-test
z_value = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_value)))

print('Sample Mean:', round(sample_mean,2))
print('Z-value:', round(z_value,3))
print('P-value:', round(p_value,4))
```

## Interpretation

- If p-value < 0.05 → Reject H₀ → Delivery time is significantly different from 30 minutes.
- If p-value > 0.05 → Accept H₀ → No strong evidence against the claim.
